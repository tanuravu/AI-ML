```python
import pandas as pd
import seaborn as sbn
data=pd.read_csv("Ecommerce Purchases")
data
```

```python
import matplotlib.pyplot as plt
french_data=data[data['Language'] =='fr']
plt.figure(figsize=(10, 10))
plt.bar(['French'], [len(french_data)], color='blue')
plt.title('Number of People with French as their Language')
plt.xlabel('Language')
plt.ylabel('Number of People')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie([len(french_data),len(data)], labels=['French','English'], autopct='%1.1f%%', colors=['blue','red'])
plt.title('Percentage of People with French as their Language')
plt.show()
```
