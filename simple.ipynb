Simple Linear Regression

```python
import seaborn as sns
```

```python
iris= sns.load_dataset('iris')
iris
```

```python
iris=iris[['petal_length','petal_width']]
iris
```

```python
X= iris['petal_length']
Y=iris['petal_width']
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
#splitttng

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.4,random_state=23)
```

```python
X_train
```

```python
Y_train
```

```python
import numpy as np
X_train=np.array(X_train).reshape(-1,1)
X_test=np.array(X_test).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr
```

```python
lr.fit(X_train,Y_train)
```

```python
c=lr.intercept_
c
```

```python
m=lr.coef_
m
```

```python
y_pres_train= m*X_train+c
y_pres_train.flatten()
```

```python
y_pres_train1=lr.predict(X_train)
y_pres_train1.flatten()
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,y_pres_train1,color='red')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
y_pres_test1=lr.predict(X_test)
y_pres_test1.flatten()
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test,y_pres_test1,color='red')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```
