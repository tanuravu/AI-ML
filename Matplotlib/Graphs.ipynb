```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#simple example:

x=np.arange(0,10)
y= np.arange(11,21)
```

    SCATTER PLOTS:-

```python
plt.scatter(x,y,c='r')
plt.xlabel('x axis') #name of the x-axis
plt.ylabel('y axis') #name of y-axis
plt.title('Graph in 20')#title of the graph
#plt.show()
```


PLOTS:-

```python
y=x*x
plt.plot(x,y,'r--',linewidth=3,markersize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D plot graph')
```

SUBPLOTS:-

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g--')
plt.subplot(2,2,2)
plt.plot(x,y,'r.-')
plt.subplot(2,2,3)
plt.plot(x,y,'b*-')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.plot(x,y)
```

Sine plot

```python
np.pi
x=np.arange(0,3*np.pi,0.2)
y=np.sin(x)
plt.title("Sine Wave")
plt.plot(x,y,'o--')
```

Bar plots:-

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,7,12]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title("bar plot")
plt.ylabel('y axis')
plt.xlabel("x axis")
```

BOX PLOT:-

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True);
```

Piechart

```python
labels='Python','C++','ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellow','lightcoral','blue']
explode=(0.2,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1d%%',shadow=True)
```
